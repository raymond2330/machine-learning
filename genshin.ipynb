```python
# Remove images without corresponding labels

import os

# Define the folder paths
images_folder = r'catdog\datasets\test\images'
labels_folder = r'catdog\datasets\test\labels'

# Get lists of image and label file names
image_files = set(os.listdir(images_folder))
label_files = set(os.listdir(labels_folder))

# Extract base names without extensions
image_bases = {os.path.splitext(file)[0] for file in image_files}
label_bases = {os.path.splitext(file)[0] for file in label_files}

# Find images without corresponding labels
images_to_remove = image_bases - label_bases

# Remove unmatched images
for image in images_to_remove:
    image_path = os.path.join(images_folder, f"{image}.jpg")  # Adjust extension as needed
    if os.path.exists(image_path):
        os.remove(image_path)
        print(f"Removed: {image_path}")
```

```python
# https://chatgpt.com/c/674c4751-4850-8009-9c99-5130bd5156ce

from ultralytics import YOLO
import torch

model = YOLO('yolo11m.pt')

# Initialize the total time complexity
total_time_complexity = 0

# Iterate over each layer in the model and calculate the time complexity
for i, layer in enumerate(model.model.modules()):
    if isinstance(layer, torch.nn.Conv2d):  # Convolutional layers only
        M_l = 32  # Example, replace with actual input size
        K_l = layer.kernel_size[0]  # Kernel size (assumes square kernel)
        C_l_1 = layer.in_channels  # Input channels from previous layer
        C_l = layer.out_channels  # Output channels of the current layer
        
        # Time complexity for the current layer
        layer_time_complexity = M_l**2 * K_l**2 * C_l_1 * C_l
        
        # Add to total complexity
        total_time_complexity += layer_time_complexity

print(f"Total Time Complexity: O({total_time_complexity})")

```
